```python
import pandas as pd
df = pd.read_csv('AttendanceMarks.csv')
df.head()
```

```python
print("statistika kehadiran:")
print(df['Attendance'].describe())
print("\nStatistika Nilai inggris:")
print(df['ESE'].describe())
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
df = df.dropna(subset=['Attendance', 'ESE'])
x = df[['Attendance']]
y = df['ESE']
model = LinearRegression()
model.fit(x, y)
beta_0 = model.intercept_
beta_1 = model.coef_[0]
print(f"Persamaan Regresi:Nilai = {beta_0:.2f} + {beta_1:.2f} * Kehadiran")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.regplot(x='Attendance', y='ESE', data=df, ci=None, line_kws={"color": "green"})
plt.title('Hubungan Kehadiran dan Nilai ESE', fontsize=16)
plt.xlabel('kehadiran (%)', fontsize=12)
plt.ylabel('Nilai ESE', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.annotate(f'Nilai = {beta_0:.2f} + {beta_1:.2f} x Kehadiran',
             xy=(0.05, 0.9),xycoords='axes fraction', fontsize=12, color='blue')
plt.show()
```
